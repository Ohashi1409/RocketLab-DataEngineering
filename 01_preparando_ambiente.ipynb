# PREPARANDO O AMBIENTE DA ATIVIDADE 🚀
- O presente notebook tem por objetivo preparar o ambiente que será utilizado para o desenvolvimento da atividade. 
- Essa etapa será responsável por :
  - Criar o catálogo que será utilizado `(medalhao_at)`
  - Criar os schemas de cada layer 
    - `bronze`
    - `silver`
    - `gold`
  - Criar o volume da **Landing**, que será nossa área de armazenamento dos arquivos crus, no formato que vierem 
- Essa etapa é de suma importância pro funcionamento das próximas etapas, uma vez que ela é a responsável por deixar pronto tudo que será usado posteriormente


## 1️⃣ CRIANDO O CATÁLOGO


```python
%sql
-- Use esse comando APENAS se for a primeira vez rodando o pipeline
-- DROP CATALOG IF EXISTS medalhao_at CASCADE;
```

```python
%sql
-- Criando o catálogo que será usado ao longo da atividade
CREATE CATALOG IF NOT EXISTS medalhao_at;    
```

## 2️⃣ CRIANDO SCHEMAS

```python
%sql
-- Usando o catálogo para criar os schemas de cada layer 
USE CATALOG medalhao_at;

-- Criando os chemas 
CREATE SCHEMA IF NOT EXISTS bronze;
CREATE SCHEMA IF NOT EXISTS silver;
CREATE SCHEMA IF NOT EXISTS gold;
```

## 3️⃣ CRIANDO VOLUME DA LANDING

```python
%sql
-- Criamos o volume dentro da pasta default
USE SCHEMA default;

-- Criando volume 
CREATE VOLUME IF NOT EXISTS landing;
```
